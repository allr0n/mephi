```python
from sympy import *
import matplotlib.pyplot as plt
%matplotlib inline
init_printing()
import numpy as np

# Максимальные рассматриваемые значения переменных x и y
# подбирается для каждой из функций индивидуально
scale = 17
ratioxy = 2 

ymax = scale
xmax = scale*ratioxy


# Диапазон [-x(y)max, x(y)max] определяет ряд параметров,
# связанных с изображением стрелок на осях координат
# и величинами смещений подписей
head_lenx = ymax/40
head_leny = xmax/40


# Текущий рисунок
fig = plt.gcf()
# Размер рисунка в дюймах
fig.set_size_inches(15, 15)
# Область рисования
ax = plt.gca()

# Интервалы изменения данных
ax.set_xlim(-xmax, xmax)
ax.set_ylim(-ymax, ymax)

# Убираем стандартные деления (и подписи) на 
# границах области данных (будущих осях)
ax.set_xticks([])
ax.set_yticks([])

# Преобразуем границы области данных:
# рисуем стрелочки и подписываем оси координат
ax.spines['left'].set_position(('data', 0))
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['top'].set_color('none')
ax.arrow(-xmax, 0.0, 2*xmax, 0.0, head_width=head_lenx,
    head_length=head_leny, overhang = 0.5, color='k',
    length_includes_head= True) 
ax.arrow(0.0, -ymax, 0.0, 2*ymax, head_width=head_leny,
    head_length=head_lenx, overhang = 0.5, color='k',
    length_includes_head= True) 
ax.text(1.0, 0.53, '$X$', fontsize=14, transform=ax.transAxes, 
    verticalalignment='center', horizontalalignment='right')
ax.text(0.47, 1.0, '$Y$', fontsize=14, transform=ax.transAxes, 
    verticalalignment='top', horizontalalignment='center')

# Вводим уравнеие исследуемой функции
x = np.linspace(-xmax, xmax, 1000)
y = ((-x+4)*(x-2)*(x+7)**2)/((x**2)*(x-6)**2)
plt.plot(x, y, 'b-')

# Описание фукнции и ее "особенных" точек
var('x')
ax.text(0, 1.0, f'Функция $y = {latex(((-x+4)*(x-2)*(x+7)**2)/((x**2)*(x-6)**2))}$', fontsize=14,
    transform=ax.transAxes, color='b', verticalalignment='top')
ax.text(0, 0.94, 'Асимптоты:', fontsize=12,
    transform=ax.transAxes, color='r', verticalalignment='top')
ax.text(0.05, 0.92, 'x = 0' , fontsize=12,
    transform=ax.transAxes, color='r', verticalalignment='top')
ax.text(0.05, 0.90, 'x = 6' , fontsize=12,
    transform=ax.transAxes, color='r', verticalalignment='top')
ax.text(0.05, 0.88, 'y = – 1' , fontsize=12,
    transform=ax.transAxes, color='r', verticalalignment='top')
ax.text(0, 0.84, 'Экстремумы:', fontsize=12,
    transform=ax.transAxes, color='k', verticalalignment='top')
ax.text(0.05, 0.82, '(– 7 , 0)' , fontsize=12,
    transform=ax.transAxes, color='k', verticalalignment='top')
ax.text(0.05, 0.80, '(3.12 , 1.25)' , fontsize=12,
    transform=ax.transAxes, color='k', verticalalignment='top')


# Подписываем координаты "особенных" точек
ax.text(-7, 0.75, '$-7.00$' , fontsize=10, color='k', verticalalignment='top', horizontalalignment='center')
ax.text(3.12, -0.3, '$3.12$' , fontsize=10, color='k', verticalalignment='top', horizontalalignment='center')
ax.text(-0.2, 1.5, '$1.25$' , fontsize=10, color='k', verticalalignment='top', horizontalalignment='right')
# ax.text(-0.2, -0.4, '$-1.00$' , fontsize=10, color='k', verticalalignment='top', horizontalalignment='right')
ax.text(6.2, 0.75, '$6.00$' , fontsize=10, color='k', verticalalignment='top', horizontalalignment='left')
ax.text(0.2, 0.75, '$0.00$' , fontsize=10, color='k', verticalalignment='top', horizontalalignment='left')


# Отображаем дополнительные линии на графике
ax.vlines(6, -ymax, ymax, color = 'r', linewidth = 2, linestyle = '--')
ax.hlines(-1, -xmax, xmax, color = 'r', linewidth = 2, linestyle = '--')
ax.vlines(3.12, 0, 1.25, color = 'k', linewidth = 2, linestyle = ':')
ax.hlines(1.25, 0, 3.12, color = 'k', linewidth = 2, linestyle = ':')


plt.show()
```
